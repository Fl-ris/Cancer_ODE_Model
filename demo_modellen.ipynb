```python
import matplotlib.pyplot as plt
from modellen import tumor_growth_models

# Plotfunctie voor enkel model
def my_plot(Ts, Vs, title="Tumor model"):
    plt.figure(figsize=(8,5))
    plt.plot(Ts, Vs, color='darkmagenta', linestyle='-', marker='o', markevery=10, label='Model')
    plt.xlabel('Tijd')
    plt.ylabel('Volume')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

# Instellingen simulatie
volume0 = 1
delta_t = 1.0
n_steps = 120

# Maak model-object
testTumor = tumor_growth_models(volume=volume0, delta_t=delta_t, n=n_steps)

# 1. Lineair
Ts, Vs = testTumor.lineaire_model(c=30)
my_plot(Ts, Vs, title="Lineair Model")

# 2. Exponentieel
Ts, Vs = testTumor.exponentieel_model(c=0.05)
my_plot(Ts, Vs, title="Exponentieel Model")

# 3. Logistisch
Ts, Vs = testTumor.logistisch_model(c=0.0005, V_max=5000)
my_plot(Ts, Vs, title="Logistisch Model")

# 4. Mendelsohn
Ts, Vs = testTumor.mendelsohn_model(c=0.001, d=1.1)
my_plot(Ts, Vs, title="Mendelsohn Model")

# 5. Montroll
Ts, Vs = testTumor.montroll_model(c=1e-6, d=1.5, V_max=5000)
my_plot(Ts, Vs, title="Montroll Model")

# 6. Allee-effect
Ts, Vs = testTumor.allee_model(c=0.00001, V_min=0.5, V_max=5000)
my_plot(Ts, Vs, title="Allee Model")

# 7. Lineair gelimiteerd
Ts, Vs = testTumor.lineair_gelimiteerd_model(c=10, d=500)
my_plot(Ts, Vs, title="Lineair Gelimiteerd Model")

# 8. Oppervlakte-gelimiteerd
Ts, Vs = testTumor.oppervlakte_gelimiteerd_model(c=30, d=500)
my_plot(Ts, Vs, title="Oppervlakte-gelimiteerd Model")

# 9. Von Bertalanffy
Ts, Vs = testTumor.von_bertalanffy_model(c=1, d=0.01)
my_plot(Ts, Vs, title="Von Bertalanffy Model")

# 10. Gompertz
Ts, Vs = testTumor.gompertz_model(c=0.05, V_max=5000)
my_plot(Ts, Vs, title="Gompertz Model")

# 11. Exponentieel afvlakkend
Ts, Vs = testTumor.exponentieel_afvlakkend_model(c=0.05, V_max=5000)
my_plot(Ts, Vs, title="Exponentieel Afvlakkend Model")
```

```python
import matplotlib.pyplot as plt
from modellen import tumor_growth_models

# Plot-functie
def my_plot(Ts, Vs, data_ts=None, data_vs=None, title="Tumor model"):
    """Plot model en optioneel meetdata"""
    plt.figure(figsize=(8,5))
    if data_ts is not None and data_vs is not None:
        plt.plot(data_ts, data_vs, 'ko', label='Data')
    plt.plot(Ts, Vs, 'purple', linestyle='-', marker='o', markevery=10, label='Model')
    plt.xlabel('Tijd')
    plt.ylabel('Volume')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()

# Gegeven data
data_ts = [0, 13, 20, 32, 42, 55, 65, 75, 85, 88, 95, 98, 107, 115, 120]
data_vs = [250, 255, 550, 575, 576, 800, 1050, 1250, 1750, 2000, 2550, 2750, 3000, 3500, 4000]

# Maak model-object
testTumor = tumor_growth_models(volume=data_vs[0], delta_t=1, n=120)

# Simulatie en plotten voor elk model
Ts, Vs = testTumor.lineaire_model(c=30)
my_plot(Ts, Vs, data_ts, data_vs, title="Lineair Model")

Ts, Vs = testTumor.exponentieel_model(c=0.05)
my_plot(Ts, Vs, data_ts, data_vs, title="Exponentieel Model")

Ts, Vs = testTumor.logistisch_model(c=0.0005, V_max=5000)
my_plot(Ts, Vs, data_ts, data_vs, title="Logistisch Model")

Ts, Vs = testTumor.mendelsohn_model(c=0.001, d=1.1)
my_plot(Ts, Vs, data_ts, data_vs, title="Mendelsohn Model")

Ts, Vs = testTumor.montroll_model(c=1e-6, d=1.5, V_max=5000)
my_plot(Ts, Vs, data_ts, data_vs, title="Montroll Model")

Ts, Vs = testTumor.allee_model(c=0.00001, V_min=100, V_max=5000)
my_plot(Ts, Vs, data_ts, data_vs, title="Allee Model")

Ts, Vs = testTumor.lineair_gelimiteerd_model(c=10, d=500)
my_plot(Ts, Vs, data_ts, data_vs, title="Lineair Gelimiteerd Model")

Ts, Vs = testTumor.oppervlakte_gelimiteerd_model(c=30, d=500)
my_plot(Ts, Vs, data_ts, data_vs, title="Oppervlakte-gelimiteerd Model")

Ts, Vs = testTumor.von_bertalanffy_model(c=1, d=0.01)
my_plot(Ts, Vs, data_ts, data_vs, title="Von Bertalanffy Model")

Ts, Vs = testTumor.gompertz_model(c=0.05, V_max=5000)
my_plot(Ts, Vs, data_ts, data_vs, title="Gompertz Model")

Ts, Vs = testTumor.exponentieel_afvlakkend_model(c=0.05, V_max=5000)
my_plot(Ts, Vs, data_ts, data_vs, title="Exponentieel Afvlakkend Model")
```
